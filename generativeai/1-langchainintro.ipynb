### Streaming And Batch

## Streaming
### Most models can stream their output content while it is being generated. By displaying output progressively, streaming significantly imporves user experience, particularly for longer responses. Calling stream() return as iterator that yields output chunks as they are produced. You can use a loop to process each chunk in real-time.

```python
model.invoke("Write me a 500 words paragraph on AI")
```

## Batch
## Batching a collection of independent requests to a model can significantly improve performance and reduce costs, as the processing can be done in parallel.

```python
responses = model.batch(
    [
        "why do parrots have colorful feathers?",
        "How do airplane fly?",
        "What is quantum computing?"
    ]
)
for response in responses:
    print(response)
```

```python
responses = model.batch(
    [
        "why do parrots have colorful feathers?",
        "How do airplane fly?",
        "What is quantum computing?"
    ]
)
config={
    'max_concurrency': 5,
    'timeout': 60,
    'streaming': True,
}
```
